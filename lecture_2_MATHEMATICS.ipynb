```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=np.array([2,3])
b=np.array([4,4])
a_dot_b= np.dot(a,b)
print(a_dot_b)
```

```python
c=np.array([40,20,35])
d=np.array([53,24,68])
c_dot_d=np.dot(c,d)
print(c_dot_d)
```

```python
# cross product of 2 vectors
a=np.array([2,3])
b=np.array([4,4])
a_cross_b= np.cross(a,b)
print(a_cross_b)
```

```python
c=np.array([40,20,35])
d=np.array([53,24,68])
c_cross_d=np.cross(c,d)
print(c_cross_d)
```

```python
# projection of a one vector on another vector
a=np.array([2,5])
v=np.array([8,-6])

magnitude=np.sqrt(sum(v**2))
projection_of_a_on_v = (np.dot(a,v)/magnitude**2)*v
```

```python
print(projection_of_a_on_v)
```

```python
import numpy as np

u = np.array([1, 2, 3])   # vector u
v = np.array([5, 6, 2])   # vector v:

# Task: Project vector u on vector v

# finding norm of the vector v
v_norm = np.sqrt(sum(v**2))

# Apply the formula as mentioned above
# for projecting a vector onto another vector
# find dot product using np.dot()
proj_of_u_on_v = (np.dot(u, v)/v_norm**2)*v

print("Projection of Vector u on Vector v is: ", proj_of_u_on_v)
```
